```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx_query import search_nodes, search_edges

G=nx.Graph()
G.add_node(1, country='RUS')
G.add_node(2, country='CHI')
G.add_node(3, country='ESP')
G.add_node(4, country='RUS')
G.add_node(5, country='BRA')
G.add_node(6, country='BRA')

G.add_edge(1,2, wire=300)
G.add_edge(1,3, wire=200)
G.add_edge(2,3, wire=200)
G.add_edge(2,4, wire=200)
G.add_edge(2,5, wire=400)
G.add_edge(6,5, wire=450)


pos = nx.spring_layout(G)
plt.figure(3,figsize=(10,10))
nx.draw(G, pos, with_labels=True, node_size=2000)
edge_labels = nx.get_edge_attributes(G,'wire')
nx.draw_networkx_edge_labels(G, pos, edge_labels)
plt.show()
```

```python

H = nx.Graph(((u, v, e) for u,v,e in G.edges(data=True) if e['wire'] >= 300))
pos = nx.spring_layout(G)
nx.draw(H, pos, with_labels=True, node_size=3000)
edge_labels = nx.get_edge_attributes(H, 'wire')
nx.draw_networkx_edge_labels(H, pos, edge_labels)

plt.show()
```
